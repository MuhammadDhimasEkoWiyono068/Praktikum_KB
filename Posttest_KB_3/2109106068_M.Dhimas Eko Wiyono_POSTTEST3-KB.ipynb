# Import Library

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Membaca data dari file CSV ke dalam DataFrame

```python
df = pd.read_csv("games-data.csv")
```

# Membuat Variabel Platform

```python
Platform = df["platform"].value_counts()
```

# Pie Chart/Diagram Lingkaran

```python
# Membuat pie chart(diagram lingkaran)
plt.pie(Platform, labels=None, startangle=90, autopct='%.2f%%')

# Mengambil label indeks (nama platform) dan hasil persen dari Pie chart
labels = [f'{index}: {percentage:.2f}%' for index, percentage in zip(Platform.index, (Platform / Platform.sum()) * 100)]

# Menambahkan legend dengan label yang berisi indeks dan hasil persen
# Kenapa Pakai legend? Karena label nya tertumpuk sehingga tidak terlihat tulisannya
plt.legend(labels, loc="best")
plt.legend(labels, bbox_to_anchor=(1.2, 1.1))

# Menambahkan judul (title) pada diagram
plt.title("Distribusi Game Berdasarkan Platform (1995-2020)")

plt.show()
```

**Penjelasan:**

Diagram lingkaran tersebut menunjukkan proporsi persentase game yang dirilis pada berbagai platform dari tahun 1995 hingga 2020. Contohnya, sekitar 25.59% dari semua game dirilis di "PC", sementara "PlayStation4" menyumbang sekitar 10.74%. Ini memberikan gambaran tentang seberapa besar kontribusi masing-masing platform terhadap total game yang dirilis selama periode tersebut. Semakin besar bagian dari lingkaran, semakin besar kontribusinya.

# HeatMap

```python
# Menghitung matriks korelasi antara kolom-kolom numerik dalam DataFrame
games_data_corr = df.corr()

# Membuat heatmap
sns.heatmap(games_data_corr, cmap="coolwarm", annot=True)

# Menambahkan judul (title) pada heatmap
plt.title("Korelasi Antar Kolom Bertipe Data Numerik")
plt.show()
```

**Penjelasan:**

Dapat melihat bahwa korelasi antara "score" dan "users" adalah sekitar 0.12, yang merupakan korelasi yang lemah. Korelasi antara "score" dan "critics" adalah sekitar 0.31, yang juga merupakan korelasi yang lemah. Korelasi antara "critics" dan "users" adalah sekitar 0.23, juga merupakan korelasi yang lemah. Tidak ada korelasi yang sangat kuat antara variabel-variabel ini berdasarkan heatmap ini.

# Bar Chart/Diagram Batang

```python
# Menghitung rata-rata data berdasarkan kolom "platform"
mean_data = df.groupby("platform").mean()

# Mengambil indeks dari mean_data, yang berisi nama-nama platform
x = mean_data.index

# Mengambil nilai rata-rata dari kolom "users" dari mean_data
y = mean_data["users"]

# Membuat diagram batang (bar chart) dengan platform di sumbu x dan rata-rata user di sumbu y
plt.bar(x, y, color="red")
plt.xticks(rotation=90)
plt.xlabel("Platform")
plt.ylabel("Mean (Rata-rata) User")

# Menambahkan judul (title) pada heatmap
plt.title("Rata-rata Jumlah Pengguna (Users) Berdasarkan Platform)")

plt.show()
```

**Penjelasan:**

Diagram Batang tersebut menunjukkan rata-rata jumlah pengguna berdasarkan platform. Dalam diagram tersebut, setiap platform memiliki nilai rata-rata pengguna yang berbeda. Contohnya, rata-rata pengguna dari platform "PlayStation4" itu hampir 400 orang, sementara rata-rata pengguna dari platform "PC" itu sekitar 260 orang. Diagram ini dapat membantu memahami sebaran relatif ukuran pengguna antar platform game. Beberapa platform memiliki rata-rata pengguna yang tinggi, sedangkan yang lainnya memiliki rata-rata pengguna yang lebih rendah. Semakin tinggi diagram batangnya, semakin banyak rata-rata penggunanya.

# Histogram

```python
# Membuat histogram dari kolom "score" dalam DataFrame df
sns.histplot(df["score"], bins=50, color="red")
plt.xlabel("Score")
plt.ylabel("Frekuensi Kemunculan")

# Menambahkan judul (title) pada histogram
plt.title("Histogram Score")

plt.show()
```

**Penjelasan:**

Histogram tersebut menunjukkan frekuensi kemunculan score dari data. Dari rentan score 0 hingga 60 terjadi berbagai kenaikan dan penurunan, tapi setelah mengalami penurunan pasti terjadi kenaikan yang lebih tinggi dari kenaikan sebelumnya. Di rentan 60-80 pola serupa terjadi, tetapi setelah mendekati 80 kenaikan yang dialami setelah mengalami penurunan itu tidak lebih tinggi dari kenaikan sebelumnya. Dan di rentan 80-100 terjadi penurunan frekuensi kemunculan yang drastis.
